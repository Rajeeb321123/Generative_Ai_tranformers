```python
# working of bpe
text = "बागमती नदी नेपालको काठमाडौँ उपत्यका भएर बगेर महत्त्वपूर्ण काठमाडौँ र ललितपुर महानगरपालिकाहरू छुट्ट्याउँदै दक्षिणी नेपालको मधेश प्रदेश हुँदै भारतको बिहार राज्यमा कमला नदीमा मिसिन्छ। यसलाई हिन्दु र बौद्ध दुवै धर्मावलम्बीले पवित्र मानेका छन्। यसको किनारमा धेरै हिन्दु मन्दिरहरू अवस्थित छन्। बागमतीको महत्व यस तथ्यमा पनि निहित छ कि हिन्दुहरूलाई यस पवित्र नदीको किनारमा दाहसंस्कार गरिन्छ। नेपाली हिन्दु परम्पराअनुसार दाहसंस्कार गर्नुअघि शवलाई बागमतीमा तीन पटक डुबाउनुपर्छ, जसले गर्दा पुनर्जन्मचक्रको अन्त्य हुन्छ। चितामा आगो बाल्ने मुख्य शोकसन्तप्त (सामान्यतया जेठो छोरा) ले दाहसंस्कार पछि तुरुन्तै पवित्र नदी-पानीले नुहाउनुपर्छ। अन्त्येष्टिमा सामेल हुने धेरै नातेदारहरू पनि नदीमा स्नान गर्छन् वा दाहसंस्कारको अन्त्यमा आफ्नो शरीरमा पवित्र जल छर्किन्छन्। बागमती नदीले मानिसलाई आध्यात्मिक तवरमा शुद्ध तुल्याउँछ भन्ने मान्यता छ।"

tokens = text.encode("utf-8")
tokens = list(map(int, tokens)) #convert to list of integers in range 0...255 for convenience
print('---')
print(text)
print('---')
print(tokens)
print("length",len(tokens))
```

```python
def get_stats(ids):
  counts = {}
  for pair in zip(ids, ids[1:]):
    counts[pair] = counts.get(pair,0) + 1
  return counts

stats = get_stats(tokens)
print(stats)
print('----')
print(sorted(((v,k) for k,v in stats.items()), reverse = True)) # sorting with v in comprehension
```

```python
def get_stats(ids):
  counts = {}
  for pair in zip(ids, ids[1:]):
    counts[pair] = counts.get(pair,0) + 1
  return counts


def merge(ids, pair, idx):
  newids = []
  i = 0
  while i < len(ids):
    if i < len(ids) - 1 and ids[i] == pair[0] and ids[i +1] == pair[1]:
      newids.append(idx)
      i += 2
    else:
      newids.append(ids[i])
      i += 1
  return newids

#-----------------------
vocab_size = 276 # final desird final vocab size
num_merges = vocab_size-256 # 20 merges . 256 is for raw bytes
ids = list(tokens)

merges = {}
for i in range(num_merges):
  stats = get_stats(ids)
  pair = max(stats, key = stats.get) # for top_pair
  idx = 256 + i
  print(f"merging {pair} into a new token {idx}")
  ids = merge(ids, pair, idx)
  merges[pair] = idx
```

**decoding**

Give a sequence of integers in the range[0, vocab_size,], what is the text?

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
print(vocab)

for (p0, p1), idx in merges.items(): # vocabulary with new added
  vocab[idx] = vocab[p0] + vocab[p1]

print(vocab)

def decode(ids):
  #given ids (list of integers), return Python string
  tokens = b"".join(vocab[idx] for idx in ids)
  text = tokens.decode("utf-8", errors='replace')
  return text

print(decode([128]))
```

encoding

The other way around: Give a string, What are the tokens?

```python
def encode(text):
  # give a string, return list of integers (the tokens)
  tokens = list(text.encode("utf-8"))
  while len(tokens) >= 2:
    stats = get_stats(tokens)
    pair = min(stats, key=lambda p: merges.get(p, float('inf')))
    if pair not in merges:
      break # nothing else can be merged
    idx = merges[pair]
    tokens = merge(tokens, pair, idx)

  return tokens

text = "बागमती नदी नेपालको काठमाडौँ उपत्यका भएर बगेर महत्त्वपूर्ण काठमाडौँ र ललितपुर महानगरपालिकाहरू छुट्ट्याउँदै दक्षिणी नेपालको मधेश प्रदेश हुँदै भारतको बिहार राज्यमा कमला नदीमा मिसिन्छ। यसलाई हिन्दु र बौद्ध दुवै धर्मावलम्बीले पवित्र मानेका छन्। यसको किनारमा धेरै हिन्दु मन्दिरहरू अवस्थित छन्। बागमतीको महत्व यस तथ्यमा पनि निहित छ कि हिन्दुहरूलाई यस पवित्र नदीको किनारमा दाहसंस्कार गरिन्छ। नेपाली हिन्दु परम्पराअनुसार दाहसंस्कार गर्नुअघि शवलाई बागमतीमा तीन पटक डुबाउनुपर्छ, जसले गर्दा पुनर्जन्मचक्रको अन्त्य हुन्छ। चितामा आगो बाल्ने मुख्य शोकसन्तप्त (सामान्यतया जेठो छोरा) ले दाहसंस्कार पछि तुरुन्तै पवित्र नदी-पानीले नुहाउनुपर्छ। अन्त्येष्टिमा सामेल हुने धेरै नातेदारहरू पनि नदीमा स्नान गर्छन् वा दाहसंस्कारको अन्त्यमा आफ्नो शरीरमा पवित्र जल छर्किन्छन्। बागमती नदीले मानिसलाई आध्यात्मिक तवरमा शुद्ध तुल्याउँछ भन्ने मान्यता छ।"
print(decode(encode(text)))
print(decode(encode(text)) == text) ## check encoding and decoding for Nepali is same
```

```python

def get_training_corpus():
    for i in range(0, len(dataset), 1000):
        yield dataset[i : i + 1000]["text"]
```

```python
from transformers import AutoTokenizer

old_tokenizer = AutoTokenizer.from_pretrained("unsloth/Qwen2-0.5B-Instruct")
```

```python
tokenizer = old_tokenizer.train_new_from_iterator(get_training_corpus(), 30000)
```

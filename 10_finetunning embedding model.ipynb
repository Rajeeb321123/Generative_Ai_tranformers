```python
from datasets import load_dataset
dataset = load_dataset("universalml0/Nepali_embedding_dataset", token="hf_",split = "train")
```

```python
dataset
```

```python
dataset.to_pandas()
```

```python
model= SentenceTransformer('intfloat/multilingual-e5-large-instruct')
```

```python
from sentence_transformers import SentenceTransformerTrainer, SentenceTransformerTrainingArguments
from sentence_transformers.losses import MultipleNegativeRankingLoss
from sentece_transformers.training_args import BatchSamplers
```

```python
loss = MultipleNegativeRankingLoss(model=model)
```

```python
args = SentenceTransformerTrainingArguments(
    output_dir = "embedding-e5-large-output",
    per_device_train_batch_size = 4,
    num_train_epochs = 2,
    warmup_ratio = 0.3,
    bf16=True,
    fp16=False,
    batch_sampler = BatchSamplers.NO_DUPLICATES,
    save_steps=1000,
    logging_steps=100,
    run_name = "embedding-e5-large-output_finetunning",
    report_to = "wandb",
)
```

```python
trainer = SentenceTransformerTrainer(
    model=model,
    args=args,
    train_dataset=dataset,
    loss=loss,
)
```

```python
trainer.train()
```
